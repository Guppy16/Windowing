```python
import scipy.io     # To Load data from matlab files: https://docs.scipy.org/doc/scipy/reference/tutorial/io.html
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Read data
mat = scipy.io.loadmat('./Data/A4-SampleData/sine5_F0A1_300Hz')
mat
```

```python
# Plot i/p and o/p data
# Seems like ch1 is Force Transducer, while ch2 is response from a floor (more details at https://www.vle.cam.ac.uk/mod/book/view.php?id=154922&chapterid=57851)
# Time is measured for 15 sec

fig, axs = plt.subplots(nrows=2, sharex=True)
t = np.linspace(0,15, num=len(mat['data_ch1']))

axs[0].plot(t, mat['data_ch1'], label="Ch1")
axs[0].set_ylabel("Amplitude")
axs[0].set_title("Force Transducer")

axs[1].plot(t, mat['data_ch2'], label="Ch2")
axs[1].set_ylabel("Amplitude")
axs[1].set_xlabel("Time / s")
axs[1].set_title("output??")

fig.suptitle("Signals for Random 20")
```

```python
# DFT

sp = np.fft.fft(np.ndarray.flatten(mat['data_ch1']))
freq = np.fft.fftfreq(len(mat['data_ch1']))

plt.plot(freq, sp.real) # NOTE: Could also plot sp.imag, but it shd be the same as the real part??
```
